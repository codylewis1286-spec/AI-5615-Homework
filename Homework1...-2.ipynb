```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

# Drop any NaNs in the relevant columns
df_fit = df[["age", "thalach"]].dropna()

X = df_fit[["age"]]      # Predictor (2D for sklearn)
y = df_fit["thalach"]    # Response (1D)

model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope: {slope:.3f}")
print(f"Intercept: {intercept:.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x="age", y="thalach", data=df, alpha=0.7)

# Regression line
x_line = np.linspace(df["age"].min(), df["age"].max(), 100)
y_line = model.predict(x_line.reshape(-1, 1))
plt.plot(x_line, y_line, color="red", linewidth=2, label="Regression line")

plt.xlabel("Age (years)")
plt.ylabel("Maximum Heart Rate (thalach)")
plt.title("Max Heart Rate vs Age")
plt.legend()
plt.show()
```

```python
# Split data
df_hd    = df.loc[df["hd"] == True,  ["age","thalach"]].dropna()
df_nohd  = df.loc[df["hd"] == False, ["age","thalach"]].dropna()

X_hd,   y_hd   = df_hd[["age"]],   df_hd["thalach"]
X_nohd, y_nohd = df_nohd[["age"]], df_nohd["thalach"]

lin_hd   = LinearRegression().fit(X_hd, y_hd)
lin_nohd = LinearRegression().fit(X_nohd, y_nohd)

slope_hd, intercept_hd     = lin_hd.coef_[0], lin_hd.intercept_
slope_nohd, intercept_nohd = lin_nohd.coef_[0], lin_nohd.intercept_

print("HD group:")
print(f"  slope = {slope_hd:.3f}, intercept = {intercept_hd:.3f}, R^2 = {lin_hd.score(X_hd, y_hd):.3f}")
print("No HD group:")
print(f"  slope = {slope_nohd:.3f}, intercept = {intercept_nohd:.3f}, R^2 = {lin_nohd.score(X_nohd, y_nohd):.3f}")

# Combined scatter and both regression lines
plt.figure(figsize=(7,5))
sns.scatterplot(data=df_nohd, x="age", y="thalach", color="#1f77b4", alpha=0.7, label="No HD")
sns.scatterplot(data=df_hd,   x="age", y="thalach", color="#d62728", alpha=0.7, label="HD")

age_min = min(df_nohd["age"].min(), df_hd["age"].min())
age_max = max(df_nohd["age"].max(), df_hd["age"].max())
x_line = pd.DataFrame({"age": np.linspace(age_min, age_max, 100)})

plt.plot(x_line["age"], lin_nohd.predict(x_line), linewidth=2, label="Fit (No HD)", color="#1f77b4")
plt.plot(x_line["age"], lin_hd.predict(x_line),   linewidth=2, label="Fit (HD)",    color="#d62728")

plt.xlabel("Age (years)")
plt.ylabel("Max Heart Rate (thalach)")
plt.title("thalach vs age by Heart Disease Status")
plt.legend()
plt.show()
```
